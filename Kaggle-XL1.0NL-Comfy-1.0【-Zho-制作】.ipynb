1.   左上角点击【file】里的【import notebook】将笔记导入，右侧菜单栏选择验证手机号
2.   右侧菜单栏【ACCELERATOR】选择GPU T4X2、【LANGUAGE】选择Python、【PERSISTENCE】选择Files only、【INTERNET】勾选开启，其他不用改

2.  进入之后先点击生成按钮【Queue Prompt】，确保生成没有问题之后点击【clear】清除默认工作流，然后将【SDXL系列标准工作流 -Zho-】json文件拖入界面即可。

# **3.启动器**

```python
%cd /kaggle/working/ComfyUI

!npm install -g localtunnel

import subprocess
import threading
import time
import socket
import urllib.request

def iframe_thread(port):
  while True:
      time.sleep(0.5)
      sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
      result = sock.connect_ex(('127.0.0.1', port))
      if result == 0:
        break
      sock.close()
  print("\nComfyUI finished loading, trying to launch localtunnel (if it gets stuck here localtunnel is having issues)\n")

  print("The password/enpoint ip for localtunnel is:", urllib.request.urlopen('https://ipv4.icanhazip.com').read().decode('utf8').strip("\n"))
  p = subprocess.Popen(["lt", "--port", "{}".format(port)], stdout=subprocess.PIPE)
  for line in p.stdout:
    print(line.decode(), end='')


threading.Thread(target=iframe_thread, daemon=True, args=(8188,)).start()

!python main.py --dont-print-server
```
